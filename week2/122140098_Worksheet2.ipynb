# Pendahuluan

## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

Misal detik 1 pada waveform di atas, terlihat amplitudo kecil, yang artinya suara pada audio sampel saat itu pelan. Dari waveform tersebut, kita juga dapat melihat sebuah pola teratur. Berikutnya yang bisa diidentifikasi melalui waveform adalah ZCR (Zero Crossing Rate). Kode dan visualisasi dapat dilihat di bawah ini. ZCR merupakan jumlah berapa kali sinyal audio melalui garis nol (berubah dari nilai positif ke negatif, dan sebaliknya). ZCR berfungsi untuk mengekstrak fitur sederhana tentang apakah audio cenderung bernada atau berisik/ritmis. Diperoleh data sebagai berikut:
- Rata-rata ZCR: 0.0490 (4.90%)
- Nilai minimum ZCR: 0.0186 (1.86%)
- Nilai maksimum ZCR: 0.1338 (13.38%)
- Standar deviasi ZCR: 0.0259 (2.59%)

Nilai ZCR tersebut dikategorikan sebagai rendah yang menandakan audio didominasi frekuensi rendah. Pada konteks musik, ZCR rendah bisa terjadi pada nada instrumen tonal seperti piano, gitar, biola, vokal bernada. Terdapat juga pola berulang yang mencerminkan struktur melodi/musik.

```python
# Load data
audio_path = os.path.join(os.getcwd(), "data", "audio_musik_gitar.mp3")
y, sr = librosa.load(audio_path)

# Hitung ZCR
frame_length = 2048  # Panjang frame untuk analisis
hop_length = 512     # Jarak antar frame

# Hitung ZCR menggunakan librosa
zcr = librosa.feature.zero_crossing_rate(y, frame_length=frame_length,hop_length=hop_length)

# Buat array waktu untuk sumbu x
times = librosa.times_like(zcr)

# Buat plot
plt.figure(figsize=(15, 6))

# Plot ZCR
plt.plot(times, zcr[0], label='Zero Crossing Rate', color='blue', linewidth=1)

# Tambahkan judul dan label
plt.title('Zero Crossing Rate Audio Gitar', size=14, pad=10)
plt.xlabel('Waktu (detik)', size=12)
plt.ylabel('Zero Crossing Rate', size=12)

# Tambahkan grid dan legend
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend()

# Statistik ZCR
print(f"Rata-rata ZCR: {zcr.mean():.4f}")
print(f"Nilai minimum ZCR: {zcr.min():.4f}")
print(f"Nilai maksimum ZCR: {zcr.max():.4f}")
print(f"Standar deviasi ZCR: {zcr.std():.4f}")

plt.tight_layout()
plt.show()
```

### A4. TODO: Spectrogram log-dB

**Instruksi:**
Hitung STFT dan tampilkan spectrogram dalam skala log-dB:
- Gunakan parameter standar (n_fft=1024, hop_length=256)
- Tampilkan dengan colorbar
- Label sumbu: waktu (detik) dan frekuensi (Hz)

**Analisis yang diperlukan:**
Jelaskan perbedaan informasi yang didapat dari spectrogram dibanding waveform.

Waveform dan Spektogram dapat digunakan untuk memvisualisasikan audio, tetapi informasi yang diberikan berbeda. Waveform memvisualisasikan audio dalam domain waktu dengan sumbu horizontal menunjukkan waktu dan sumbu vertikal menunjukkan amplitudo. Dari waveform, kita dapat memperoleh informasi terkait durasi, bentuk gelombang, Zero Cross Rate, serta intensitas suara, tetapi informasi mengenai frekuensi yang membentuk suara tidak terlihat. Informasi terkait frekuensi dapat diamati dengan spektogram. Spektrogram merepresentasikan audio dalam domain waktu–frekuensi, di mana sumbu horizontal menunjukkan waktu, sumbu vertikal menunjukkan frekuensi, dan warna atau tingkat kecerahan merepresentasikan intensitas energi suara. Dengan spektrogram, kita bisa melihat frekuensi dominan dari audio, perubahan nada, timbre, serta pola spektral yang muncul sepanjang waktu. Oleh karena itu, waveform lebih cocok untuk memahami kapan dan seberapa kuat suara terjadi, sedangkan spectrogram lebih cocok untuk mengetahui nada atau frekuensi apa saja yang ada di dalam suara dan bagaimana perubahannya dari waktu ke waktu.

Sebagai contoh adalah (`audio_musik_gitar.mp3`) yang saya gunakan ini. Melalui spektogram, dapat diperoleh informasi berikut:
- Intensitas suara audio terkuat berada pada range frekuensi 0-2000 Hz (warna kuning-hijau)
- Terdapat garis-garis konsisten di bagian bawah yang menunjukkan nada dasar gitar (sekitar 200-500 Hz)
- Timbre (warna khas suara) gitar dapat dikenali dengan adanya harmonik (muncul sebagai garis-garis horizontal/mendatar pada spektogram)
- Harmonik-harmonik muncul hingga sekitar 8000 Hz dengan intensitas yang semakin melemah ke 10000 Hz
- Setiap pola vertikal berjarak sekitar 0.5-1 detik, menunjukkan tempo permainan gitar
- Pola berulang vertikal muncul teratur menunjukkan ciri khas audio musik

```python
# Load audio file
audio_path = os.path.join(os.getcwd(), "data", "audio_musik_gitar.mp3")
y, sr = librosa.load(audio_path)

# Compute STFT (Short-Time Fourier Transform) with Hann window
D = librosa.stft(y, n_fft=1024, hop_length=256, window='hann')

# Convert to log-scaled DB units  
DB = librosa.amplitude_to_db(abs(D), ref=np.max)

# Calculate parameters
time_resolution = hop_length / sr  # in seconds
freq_resolution = sr / n_fft  # in Hz
magnitude_shape = D.shape
magnitude_range = (np.abs(D).min(), np.abs(D).max())

# Create figure with larger size
plt.figure(figsize=(12, 8))

# Display spectrogram
img = librosa.display.specshow(
    DB,
    sr=sr, 
    hop_length=256,
    x_axis='time',
    y_axis='hz',
    cmap='viridis'
)

# Add colorbar
cbar = plt.colorbar(img, format='%+2.0f dB')
cbar.set_label('Intensity (dB)', rotation=270, labelpad=15)

# Add title and labels
plt.title('Spectrogram (log-dB scale) Audio Musik Gitar', pad=10, size=14)
plt.xlabel('Waktu (detik)', size=12)
plt.ylabel('Frekuensi (Hz)', size=12)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
# Print information
print("STFT Parameters & Information:")
print("=" * 40)
print(f"FFT size (n_fft): {n_fft} samples")
print(f"Hop length: {hop_length} samples")
print(f"Window function: {window}")
print(f"Magnitude shape: {magnitude_shape} (frequencies × time frames)")
print(f"Time resolution: {time_resolution*1000:.2f} ms")
print(f"Frequency resolution: {freq_resolution:.2f} Hz")
print(f"Magnitude range: [{magnitude_range[0]:.2e}, {magnitude_range[1]:.2e}]")
```

### A5. TODO: MFCC

**Instruksi:**
Hitung dan tampilkan minimal 13 koefisien MFCC sebagai heatmap:
- Sumbu X: waktu (frame)
- Sumbu Y: koefisien MFCC (1-13)
- Gunakan colorbar dan judul yang jelas

**Analisis yang diperlukan:**
Interpretasi sederhana: apakah pola MFCC stabil atau berubah-ubah? Apa potensi maknanya?

Pola MFCC yang dihasilkan oleh (`audio_musik_gitar.mp3`) cenderung teratur/stabil pada koefisien 0-3, sementara pada koefisien 4-12 memiliki pola yang sedikit agak bervariasi (nyaris samar). Hal ini dapat berarti suara relatif tetap dalam jangka waktu tertentu. Kondisi ini biasanya muncul ketika timbre audio tidak mengalami banyak variasi, misalnya saat gitar dipetik lalu nadanya sustain.

```python
# Load audio file
audio_path = os.path.join(os.getcwd(), "data", "audio_musik_gitar.mp3")
y, sr = librosa.load(audio_path)

# Define parameters
n_fft = 1024
hop_length = 256
n_mfcc = 13

# Create figure with two subplots
plt.figure(figsize=(15, 12))

# Plot 1: Mel-spectrogram
plt.subplot(2, 1, 1)
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, n_fft=n_fft)
mel_spect_db = librosa.power_to_db(mel_spect, ref=np.max)
img1 = librosa.display.specshow(
    mel_spect_db,
    x_axis='time',
    y_axis='mel',
    sr=sr,
    fmax=8000,
    cmap='viridis'
)
plt.colorbar(img1, format='%+2.0f dB', label='Amplitude')
plt.title('Mel-Spectrogram (dB)', pad=10, size=14)
plt.xlabel('Waktu (detik)', size=12)
plt.ylabel('Frekuensi Mel', size=12)

# Plot 2: MFCC
plt.subplot(2, 1, 2)
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft)
img2 = librosa.display.specshow(
    mfccs,
    x_axis='time',
    y_axis='frames',
    hop_length=hop_length,
    sr=sr,
    cmap='viridis'
)
plt.colorbar(img2, format='%+2.0f', label='Magnitude')
plt.title(f'MFCC Coefficients Audio Musik Gitar ({n_mfcc} koefisien)', pad=10, size=14)
plt.xlabel('Waktu (detik)', size=12)
plt.ylabel('Koefisien MFCC', size=12)

# Adjust layout
plt.tight_layout()
plt.show()

# Print statistics
print("\nMel-spectrogram Statistics:")
print(f"Shape: {mel_spect_db.shape}")
print(f"Mean: {mel_spect_db.mean():.2f}")
print(f"Std: {mel_spect_db.std():.2f}")
print(f"Min: {mel_spect_db.min():.2f}")
print(f"Max: {mel_spect_db.max():.2f}")

print("\nMFCC Statistics:")
print(f"Shape: {mfccs.shape}")
print(f"Mean: {mfccs.mean():.2f}")
print(f"Std: {mfccs.std():.2f}")
print(f"Min: {mfccs.min():.2f}")
print(f"Max: {mfccs.max():.2f}")
```

- Label sumbu: intensitas pixel dan frekuensi

**Analisis yang diperlukan:**
Analisis: channel mana yang dominan? Bagaimana kontras gambar? Seperti apa sebaran intensitasnya?

Berdasarkan histogram RGB dari (`gambar_pemandangan_jalanan.jpg`), diperoleh:

1. Dominasi Channel
- Channel biru (Blue) tampak paling dominan di sisi kanan histogram (intensitas tinggi, >180). Artinya, gambar punya banyak area dengan warna kebiruan terang (langit).
- Channel hijau (Green) juga cukup kuat di rentang menengah (sekitar 80–160).
- Channel merah (Red) cenderung kuat di intensitas rendah–menengah (sekitar 50–120), tapi tidak sekuat hijau dan biru di bagian intensitas tinggi.

Jadi secara umum, biru mendominasi, disusul hijau, lalu merah.

2. Kontras Gambar
- Histogram menunjukkan distribusi intensitas yang menyebar merata hampir di seluruh rentang (0–255).
- Ada puncak jelas di area rendah-menengah (sekitar 80–120) dan tinggi (sekitar 200–250), menandakan ada pemisahan yang baik antara area terang dan gelap.
- Distribusi merata berarti gambar memiliki kontras yang baik.

3. Sebaran Intensitas
- Intensitas rendah (0–50): relatif sedikit, jadi area gelap tidak terlalu dominan.
- Intensitas menengah (80–150): banyak frekuensi, terutama hijau dan merah, menandakan adanya banyak objek dengan warna natural (jalanan, bangunan).
- Intensitas tinggi (200–255): sangat dominan pada biru, cukup besar juga pada hijau, ini mengindikasikan area terang seperti langit cerah atau objek bercahaya putih-kebiruan.

```python
try:
    # Read image using OpenCV
    img_path = os.path.join(os.getcwd(), "data", "gambar_pemandangan_jalanan.jpg")
    img_bgr = cv2.imread(img_path)
    
    if img_bgr is None:
        raise FileNotFoundError("Image could not be loaded")
    
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    # Create figure with larger size
    plt.figure(figsize=(15, 6))
    
    # Calculate histograms for each channel
    colors = ('red', 'green', 'blue')
    for i, color in enumerate(colors):
        hist = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
        plt.plot(hist, color=color, label=f'Channel {color.upper()}')
    
    # Customize plot
    plt.title('Histogram RGB Pemandangan Jalanan', pad=10, size=14)
    plt.xlabel('Intensitas Pixel (0-255)', size=12)
    plt.ylabel('Frekuensi', size=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Adjust layout and display
    plt.tight_layout()
    plt.show()

    # Print channel statistics
    print("\nStatistik Kanal Warna:")
    print("=" * 40)
    for i, color in enumerate(colors):
        channel = img_rgb[:,:,i]
        print(f"\nKanal {color.upper()}:")
        print(f"Mean: {channel.mean():.2f}")
        print(f"Std Dev: {channel.std():.2f}")
        print(f"Min: {channel.min()}")
        print(f"Max: {channel.max()}")

except FileNotFoundError as e:
    print(f"Error: {str(e)}")
    print("Please check if the image file exists in the data directory")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

### C2. TODO: Baca & Metadata

**Instruksi:**
Baca video dengan OpenCV dan tampilkan metadata:
- Resolusi (Width × Height)
- Frame rate (fps)
- Jumlah total frame
- Durasi (detik)
- Klasifikasi resolusi (HD, Full HD, 4K, dll.)

**Analisis yang diperlukan:**
Jelaskan pentingnya parameter-parameter tersebut untuk analisis video atau aplikasi tertentu.

Parameter-parameter tersebut sangat penting karena memengaruhi kualitas visual, detail informasi, dan beban komputasi dalam analisis maupun aplikasi. Resolusi (Width × Height) menentukan detail gambar, di mana resolusi tinggi lebih baik untuk mendeteksi objek kecil. Frame rate (fps) menunjukkan kelancaran gerakan; fps tinggi dibutuhkan untuk aplikasi real-time seperti pelacakan atau gaming. Jumlah total frame dan durasi (detik) menunjukkan panjang video—semakin panjang, semakin besar kebutuhan memori dan waktu proses. Klasifikasi resolusi (HD, Full HD, 4K, dll.) memberi standar kualitas visual; misalnya 4K cocok untuk pengawasan detail, sedangkan resolusi rendah lebih efisien untuk streaming atau analisis sederhana. Dengan memahami parameter ini, analisis bisa lebih tepat sekaligus efisien.

```python
from datetime import timedelta

try:
    # Read video file
    video_path = os.path.join(os.getcwd(), "data", "video_pemandangan_kembangapi.mp4")
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        raise FileNotFoundError("Error: Could not open video file")
    
    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps
    
    # Determine resolution classification
    def classify_resolution(w, h):
        if w >= 3840 and h >= 2160:
            return "4K (2160p)"
        elif w >= 2560 and h >= 1440:
            return "2K (1440p)"
        elif w >= 1920 and h >= 1080:
            return "Full HD (1080p)"
        elif w >= 1280 and h >= 720:
            return "HD (720p)"
        else:
            return "SD"
    
    resolution_class = classify_resolution(width, height)
    
    # Print metadata with formatting
    print("📹 VIDEO METADATA")
    print("=" * 40)
    print(f"📐 Resolusi: {width} × {height} pixels")
    print(f"⚡ Frame rate: {fps} fps")
    print(f"🎞️ Total frames: {total_frames:,}")
    print(f"⏱️ Durasi: {duration:.2f} seconds ({str(timedelta(seconds=duration))})")
    print(f"📺 Klasifikasi: {resolution_class}")
    
    # Release video capture
    cap.release()

except FileNotFoundError as e:
    print(f"Error: {str(e)}")
    print("Please check if the video file exists in the data directory")
except Exception as e:
    print(f"An error occurred: {str(e)}")
    if 'cap' in locals():
        cap.release()
```

**Analisis yang diperlukan:**

**Perbedaan visual:**

- Percikan cahaya sebelumnya telah menghilang total, menyisakan suasana gelap, dan pencahayaan hanya dari lampu bangunan

**Yang dapat dipelajari:**

Dari cuplikan frame video tersebut, ada beberapa hal yang bisa dipelajari. Dari segi waktu, video menunjukkan perubahan dari suasana gelap hingga terang selama sekitar 16 detik. Dari pencahayaan, terlihat kontras yang kuat antara bagian gelap dan terang, dengan cahaya kembang api yang tiba-tiba membuat kamera menyesuaikan pencahayaannya. Dari sisi teknis sederhana, kecepatan gambar yang cukup membantu menangkap momen ledakan, kamera harus bisa menampilkan perbedaan terang-gelap dengan baik, dan warna perlu ditampilkan sesuai aslinya agar terlihat natural.

```python
# Buka video dengan os.path.join
video_path = os.path.join(os.getcwd(), "data", "video_pemandangan_kembangapi.mp4")
cap = cv2.VideoCapture(video_path)

# Dapatkan total frame dan FPS
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Tentukan frame yang akan diambil
frames_to_extract = [
    0,                    # Frame pertama
    total_frames // 2,    # Frame tengah
    total_frames - 1      # Frame terakhir
]

# List untuk menyimpan frame dan timestamp
frames = []
timestamps = []

# Ekstrak frame yang diinginkan
for frame_idx in frames_to_extract:
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
    ret, frame = cap.read()
    if ret:
        # Konversi BGR ke RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(frame_rgb)
        # Hitung timestamp dalam detik
        timestamp = frame_idx / fps
        timestamps.append(f"{timestamp:.2f}s")

# Tutup video
cap.release()

# Tampilkan frame dalam subplot
plt.figure(figsize=(15, 5))

for idx, (frame, timestamp) in enumerate(zip(frames, timestamps)):
    plt.subplot(1, 3, idx + 1)
    plt.imshow(frame)
    plt.title(f'Frame {frames_to_extract[idx]}\nTimestamp: {timestamp}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
